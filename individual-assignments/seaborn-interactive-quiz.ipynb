```python
import pandas as pd
import matplotlib.pyplot as plt
import altair as alt

import warnings
warnings.filterwarnings('ignore')

data = pd.read_csv('onlinefoods.csv')
print(data.head())
```

```python
# Problem 1: Histogram - Customer Age Distribution

# 1. Histogram: Customer Age Distribution
age_hist = alt.Chart(data).mark_bar().encode(
    alt.X('Age:Q', bin=alt.Bin(maxbins=20), title='Age Bins'),
    y='count()',
    tooltip=['count()']
).properties(title='Customer Age Distribution').interactive()

age_hist
```

```python
# Problem 2: Bar Chart - Gender-Based Food Ordering Trends
gender_bar = alt.Chart(data).mark_bar().encode(
    x='Gender:N',
    y='count()',
    color='Gender:N',
    tooltip=['count()']
).properties(
    title='Gender-Based Food Ordering Trends',
    width=600,
    height=400
).interactive()

gender_bar
```

```python
# Problem 3: Box Plot - Family Size Distribution
data = pd.read_csv('onlinefoods.csv')
alt.Chart(data).mark_boxplot().encode(
    y=alt.Y('Family size:Q', title='Family Size')
).properties(
    title='Box Plot: Family Size Distribution',
    width=400
) .interactive()
```

```python
# Problem 4: Pie Chart - Educational Qualifications

data = data.drop(columns=["Unnamed: 12"])

edu_counts = data["Educational Qualifications"].value_counts().reset_index()
edu_counts.columns = ["Educational Qualifications", "Count"]

chart = alt.Chart(edu_counts).mark_arc(innerRadius=50).encode(
    theta=alt.Theta(field="Count", type="quantitative"),
    color=alt.Color(field="Educational Qualifications", type="nominal"),
    tooltip=["Educational Qualifications", "Count"]
).properties(
    title="Educational Qualifications of Customers",
    width=400,
    height=400
).interactive()

chart
```

```python
# Problem 5: Scatter Plot - Age vs Family Size
data = pd.read_csv('onlinefoods.csv')
alt.Chart(data).mark_circle(size=60).encode(
    x='Age:Q',
    y='Family size:Q',
    color='Gender:N',
    tooltip=['Age', 'Family size', 'Gender']
).properties(
    title='Age vs. Family Size Colored by Gender',
    width=600,
    height=400
).interactive()
```

```python
# Problem 6: Heatmap - Family Size vs Age
family_age_heatmap = alt.Chart(data).mark_rect().encode(
    alt.X('Age:O', title='Age'),
    alt.Y('Family size:O', title='Family Size'),
    alt.Color('count():Q', scale=alt.Scale(scheme='blues')),
    tooltip=['Age:O', 'Family size:O', 'count()']
).properties(title='Family Size vs Age Heatmap').interactive()

family_age_heatmap
```

```python
# Problem 7: Bar Chart - Most Common Customer Occupations
import altair as alt
import pandas as pd

occupation_bar = alt.Chart(data).mark_bar().encode(
    y=alt.Y('Occupation:N', sort='-x'),
    x='count()',
    color='Occupation:N',
    tooltip=['count()']
).properties(title='Most Common Customer Occupations').interactive()

occupation_bar
```

```python
# Problem 8: Scatter Plot - Age vs Latitude
age_lat_scatter = alt.Chart(data).mark_circle(size=60).encode(
    alt.X('latitude:Q', title='Latitude'),
    alt.Y('Age:Q', title='Age'),
    alt.Color('Family size:Q', scale=alt.Scale(scheme='viridis')),
    tooltip=['latitude:Q', 'Age:Q', 'Family size:Q']
).properties(title='Geographic Distribution: Age vs Latitude').interactive()

age_lat_scatter
```

```python
#Problem 9: Violin Plot - Family Size by Gender
family_violin = alt.Chart(data).transform_density(
    'Family size',
    as_=['Family size', 'density'],
    groupby=['Gender']
).mark_area(
    opacity=0.7,
    interpolate='basis'
).encode(
    alt.X('Family size:Q', title='Family Size'),
    alt.Y('density:Q', title='Density', stack='center'),
    alt.Color('Gender:N', scale=alt.Scale(range=['#ff7f0e', '#1f77b4'])),
    tooltip=['Gender:N', 'Family size:Q', 'density:Q']
).properties(
    title='Family Size Distribution by Gender (Violin Plot)',
    width=500,
    height=300
).interactive()


family_violin
```

```python
# Problem 10: Geographic Map - Customer Locations
geo_map = alt.Chart(data).mark_circle(size=100).encode(
    alt.X('longitude:Q', title='Longitude'),
    alt.Y('latitude:Q', title='Latitude'),
    alt.Color('Age:Q', scale=alt.Scale(scheme='viridis')),
    alt.Size('Family size:Q'),
    tooltip=['latitude:Q', 'longitude:Q', 'Age:Q', 'Gender:N', 'Family size:Q']
).properties(title='Customer Locations Map').interactive()

geo_map
```

```python
# Problem 11: Bar Chart - Satisfaction by Gender
satisfaction_gender = alt.Chart(data).mark_bar().encode(
    alt.X('Gender:N', title='Gender'),
    alt.Y('count():Q', title='Number of Customers'),
    alt.Color('Output:N', scale=alt.Scale(domain=['Yes', 'No'], range=['#2ca02c', '#d62728'])),
    tooltip=['Gender:N', 'Output:N', 'count()']
).properties(title='Customer Satisfaction by Gender').interactive()

satisfaction_gender
```

```python
# Problem 12: Box Plot - Age vs Satisfaction
age_satisfaction_box = alt.Chart(data).mark_boxplot().encode(
    alt.X('Output:N', title='Satisfaction Level'),
    alt.Y('Age:Q', title='Age'),
    alt.Color('Output:N', scale=alt.Scale(domain=['Yes', 'No'], range=['#2ca02c', '#d62728'])),
    tooltip=['Output:N', 'Age:Q']
).properties(title='Impact of Age on Customer Satisfaction').interactive()

age_satisfaction_box
```

```python
# Problem 13: Scatter Plot - Monthly Income vs Age
income_age_scatter = alt.Chart(data).mark_circle(size=80).encode(
    alt.X('Age:Q', title='Age'),
    alt.Y('Monthly Income:N', title='Monthly Income'),
    alt.Color('Educational Qualifications:N'),
    tooltip=['Age:Q', 'Monthly Income:N', 'Educational Qualifications:N', 'Occupation:N']
).properties(title='Monthly Income vs Age').interactive()

income_age_scatter
```

```python
# Problem 14: Fix for Feedback Pie Chart
feedback_pie = alt.Chart(data).mark_arc(innerRadius=50).encode(
    alt.Theta('count():Q'),
    alt.Color('Feedback:N', scale=alt.Scale(domain=['Positive', 'Negative '], range=['#2ca02c', '#d62728'])),
    tooltip=['Feedback:N', 'count()']
).properties(title='Customer Feedback Distribution')

feedback_pie
```

```python
# Problem 16: Grouped Bar Chart - Satisfaction by Occupation
satisfaction_occupation = alt.Chart(data).mark_bar().encode(
    alt.X('Occupation:N', title='Occupation'),
    alt.Y('count():Q', title='Number of Customers'),
    alt.Color('Output:N', scale=alt.Scale(domain=['Yes', 'No'], range=['#2ca02c', '#d62728'])),
    tooltip=['Occupation:N', 'Output:N', 'count()']
).properties(title='Satisfaction Levels by Occupation').interactive()

satisfaction_occupation
```

```python
# Problem 17: Heatmap - Correlation Matrix
numeric_data = data[['Age', 'Family size']].copy()
numeric_data['Satisfaction'] = data['Output'].map({'Yes': 1, 'No': 0})
corr_data = numeric_data.corr().reset_index().melt('index')
corr_data.columns = ['Variable1', 'Variable2', 'Correlation']

correlation_heatmap = alt.Chart(corr_data).mark_rect().encode(
    alt.X('Variable1:N', title='Variables'),
    alt.Y('Variable2:N', title='Variables'),
    alt.Color('Correlation:Q', scale=alt.Scale(scheme='blueorange', domain=[-1, 1])),
    tooltip=['Variable1:N', 'Variable2:N', 'Correlation:Q']
).properties(title='Correlation Matrix: Age, Family Size, Satisfaction').interactive()

correlation_heatmap
```

```python
# Problem 18: Linked Plots
brush = alt.selection_interval()

plot1 = alt.Chart(data).mark_circle(size=60).add_selection(brush).encode(
    alt.X('Age:Q', title='Age'),
    alt.Y('Family size:Q', title='Family Size'),
    alt.Color('Gender:N'),  # Simple color encoding
    opacity=alt.condition(brush, alt.value(0.8), alt.value(0.1)), 
    tooltip=['Age:Q', 'Family size:Q', 'Gender:N']
).properties(title='Age vs Family Size', width=250, height=200)

plot2 = alt.Chart(data).mark_circle(size=60).encode(
    alt.X('Age:Q', title='Age'),
    alt.Y('Output:N', title='Satisfaction'),
    alt.Color('Gender:N'),  # Simple color encoding
    opacity=alt.condition(brush, alt.value(0.8), alt.value(0.1)), 
    tooltip=['Age:Q', 'Output:N', 'Gender:N']
).properties(title='Age vs Satisfaction', width=250, height=200)


linked_plots = alt.hconcat(plot1, plot2)
linked_plots
```

```python
# Problem 19: Interactive Mean Calculation
occupation_means = data.groupby('Occupation')['Family size'].mean().reset_index()
occupation_means.columns = ['Occupation', 'Mean_Family_Size']

mean_chart = alt.Chart(occupation_means).mark_bar().encode(
    alt.X('Occupation:N', title='Occupation'),
    alt.Y('Mean_Family_Size:Q', title='Mean Family Size'),
    tooltip=['Occupation:N', 'Mean_Family_Size:Q']
).properties(title='Mean Family Size by Occupation').interactive()

mean_chart
```

```python
# Problem 20: Dynamic Bar & Heatmap - Occupation vs Age & Family Size
click = alt.selection_multi(fields=['Occupation'])

bar_chart = alt.Chart(data).mark_bar().add_selection(click).encode(
    x='count():Q',
    y=alt.Y('Occupation:N', sort='-x'),
    color=alt.value('steelblue'),
    tooltip=['Occupation:N', 'count()']
).properties(width=250, height=200)

heatmap_chart = alt.Chart(data).mark_rect().transform_filter(click).encode(
    alt.X('Age:O'),
    alt.Y('Family size:O'),
    alt.Color('count():Q', scale=alt.Scale(scheme='blues')),
    tooltip=['Age:O', 'Family size:O', 'count()']
).properties(width=350, height=200)

combined_chart = alt.hconcat(bar_chart, heatmap_chart)
combined_chart
```
